```python
import pickle
import matplotlib.pyplot as plt
```

```python
def load_losses(path: str) -> list[float]:
    with open(path, 'rb') as f:
        losses = pickle.load(f)
        train_losses = losses[0]
        val_losses = losses[1]
    return train_losses, val_losses
```

```python
rnn_train_loss, rnn_val_losses = load_losses(path='../model_results/rnn_losses.pkl')
lstm_train_loss, lstm_val_losses = load_losses(path='../model_results/lstm_losses.pkl')
transformer_train_loss, transformer_val_losses = load_losses(path='../model_results/transformer_losses.pkl')
```

```python
print(rnn_train_loss[-1], rnn_val_losses[-1])
print(lstm_train_loss[-1], lstm_val_losses[-1])
print(transformer_train_loss[-1], transformer_val_losses[-1])
```

```python
# RNN Plot
plt.figure(figsize=(6, 4))
plt.plot(rnn_train_loss, label='Training Loss')
plt.plot(rnn_val_losses, label='Validation Loss')
plt.title('RNN Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# LSTM Plot
plt.figure(figsize=(6, 4))
plt.plot(lstm_train_loss, label='Training Loss')
plt.plot(lstm_val_losses, label='Validation Loss')
plt.title('LSTM Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Transformer Plot
plt.figure(figsize=(6, 4))
plt.plot(transformer_train_loss, label='Training Loss')
plt.plot(transformer_val_losses, label='Validation Loss')
plt.title('Transformer Loss per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
